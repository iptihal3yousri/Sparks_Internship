<h3>Modeling</h3>

>split the data

```python
x = df['Hours'].values.reshape(-1,1)
y = df['Scores'].values.reshape(-1, 1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.2, random_state=22)
```

```python
Linear_Regression_model = LinearRegression()
Linear_Regression_model.fit(xtrain, ytrain)
prediction = Linear_Regression_model.predict(xtest)
```

>plot the model with the data

```python
plt.scatter(df['Hours'], df['Scores'])
plt.plot(xtest, prediction, c='black')
plt.ylabel('Scores')
plt.xlabel('Hours')
plt.show()
```

>Evaluate the Model

```python
print('The Linear Regression Model accuracy : ', Linear_Regression_model.score(xtrain, ytrain))
print('The Linear Regression Model Mean Absolute Error : ', mean_absolute_error(ytest, prediction))
print('The Linear Regression Model Mean Squared Error : ', mean_squared_error(ytest, prediction))
```
